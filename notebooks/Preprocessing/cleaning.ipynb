```python
import json
from pathlib import Path
import re
```

```python
# remove HTML and markdown artifacts, lowercase but keep names, remove non-character letters, ensure whitespace seperated words
# Beginning\n\n
# \n\nMiddle\n\n
# \n\nEnding\n\n
# jewelry—things
# \"\n\n
```

```python
def clean_story_for_embeddings(text):
    """
    Clean story text for training word embeddings.
    
    Steps:
    1. Remove HTML and Markdown artifacts.
    2. Replace long dashes (—, –) with space.
    3. Preserve contractions (e.g., didn't) and hyphenated words (e.g., human-written).
    4. Remove other non-alphanumeric characters.
    5. Normalize whitespace.
    """
    # 1. Remove HTML/Markdown
    text = re.sub(r'<[^>]+>', ' ', text)          # HTML tags
    text = re.sub(r'\[[^\]]*\]', ' ', text)       # Markdown links/references
    
    # 2. Replace long dashes with space
    text = re.sub(r'[—–]', ' ', text)
    
    # 3. Remove unwanted characters but keep letters, numbers, apostrophes, and hyphens inside words
    # We replace unwanted symbols with space
    text = re.sub(r"(?:[^a-zA-Z0-9'\- ]+)", ' ', text)
    
    # 4. Normalize whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
clean_story_for_embeddings("Mara went to the store—she bought jewelry—things like necklaces. <br> It was amazing! Joseph didn't care. He could read a human-written book")
```
