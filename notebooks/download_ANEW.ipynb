```python
from git import Repo
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path("/Users/tildeidunsloth/Desktop/Thesis")
DATA_ROOT = PROJECT_ROOT / "data" / "X-ANEW"


def get_xanew(
    split: str = "train",
    root: Path = DATA_ROOT,
    url: str = "https://github.com/JULIELab/X-ANEW.git"
):
    assert split in {"train", "dev", "test"}, 'split must be "train", "dev", or "test".'

    root = Path(root)
    root.mkdir(parents=True, exist_ok=True)

    if not (root / ".git").exists():
        print("Cloning X-ANEW dataset...")
        Repo.clone_from(url, root)


# downloads the dataset if necessary and returns a pandas DataFrame
get_xanew('train')
```
