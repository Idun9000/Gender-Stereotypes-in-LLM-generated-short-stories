### Import packages, define paths and load data

```python
import os
import json
from gensim.models.word2vec import Word2Vec
from pathlib import Path
from preprocessing import preprocess_story_for_embeddings
import re
import numpy as np
import random
```

```python
# Define project and data paths
PROJECT_ROOT = Path("/Users/tildeidunsloth/Desktop/Thesis")
DATA_DIR = PROJECT_ROOT / "data/cleaned"
EMBEDDINGS_DIR = PROJECT_ROOT / "embeddings"
```

```python
romance_data_path = DATA_DIR / "romance_stories_cleaned.jsonl"
sci_fi_data_path = DATA_DIR / "sci_fi_stories_cleaned.jsonl"
lit_fic_data_path = DATA_DIR / "lit_fiction_stories_cleaned.jsonl"
```

```python
def load_jsonl_data(file_path):
    with open(file_path, 'r') as f:
        stories = [json.loads(line) for line in f]
    return stories
```

```python
lit_fiction_stories = load_jsonl_data(lit_fic_data_path)
romance_stories = load_jsonl_data(romance_data_path)
sci_fi_stories = load_jsonl_data(sci_fi_data_path)
```

### Preprocess stories for embeddings

```python
# Preprocess stories for Word2Vec
sentences_romance = [
    sentence
    for story in romance_stories
    for sentence in preprocess_story_for_embeddings(story['story'])
]

sentences_sci_fi = [
    sentence
    for story in sci_fi_stories
    for sentence in preprocess_story_for_embeddings(story['story'])
]

sentences_lit_fic = [
    sentence
    for story in lit_fiction_stories
    for sentence in preprocess_story_for_embeddings(story['story'])
]
```

### Set Word2Vec parameters

```python
# Word2Vec parameters
num_features = 300  # Word vector dimensionality
min_word_count = 20  # Minimum word count
context = 5  # Context window size
downsampling = 1e-5  # Downsample setting for frequent words
seed = 2000
```

### Function for training embeddings

```python
def train_and_save_models(sentences, output_dir, genre, **word2vec_params):
    model = Word2Vec(sentences=sentences, **word2vec_params)
    model_name = f"{genre}.w2v"
    save_path = output_dir / model_name
    model.save(str(save_path))
```

```python
def train_and_save_models_seed(sentences, output_dir, genre, **word2vec_params):
    model = Word2Vec(sentences=sentences, **word2vec_params)
    
    seed = word2vec_params.get("seed", "")
    model_name = f"{genre}_{seed}.w2v"
    
    save_path = output_dir / model_name
    model.save(str(save_path))
```

### Train models

```python
train_and_save_models(sentences_sci_fi, 
                      EMBEDDINGS_DIR, 
                      "sci_fi", 
                      vector_size=num_features,
                      min_count=min_word_count,
                      window=context,
                      sample = downsampling,
                      sg=1,  # Use Skip-gram
                      workers=8,
                      negative=10,
                      epochs = 15,
                      seed = 2000)
```

```python
train_and_save_models(sentences_romance, 
                      EMBEDDINGS_DIR, 
                      "romance", 
                      vector_size=num_features,
                      min_count=min_word_count,
                      window=context,
                      sample = downsampling,
                      sg=1,  # Use Skip-gram
                      workers=8,
                      negative=10,
                      epochs = 15,
                      seed = 2000)
```

```python
train_and_save_models(sentences_lit_fic, 
                      EMBEDDINGS_DIR, 
                      "literary_fiction", 
                      vector_size=num_features,
                      min_count=min_word_count,
                      window=context,
                      sample = downsampling,
                      sg=1,  # Use Skip-gram
                      workers=8,
                      negative=10,
                      epochs = 15,
                      seed = 2000)
```

### Train models with different seeds for testing

```python
for seeds in [42, 123, 999]:
    params = {"vector_size": num_features, "seed": seeds, "min_count": min_word_count, "window": context, "sample": downsampling, "sg": 1, "workers": 1, "negative": 10, "epochs": 15}
    train_and_save_models_seed(sentences_romance, EMBEDDINGS_DIR, genre="romance", **params)
```
