### Define agents, events and contexts

```python
agents = {
    'occupational_masculine_stereotyped': [
        'An engineer', 'An architect', 'A pilot', 'A construction worker', 'A barber'
    ],
    'occupational_feminine_stereotyped': [
        'A nurse', 'A librarian', 'An assistant', 'primary school teacher', 'A psychologist'
    ],
    'occupational_neutral': [
        'An artist', 'A chiropractor', 'A cook', 'A coordinator', 'A musician'
    ],
    'non_occupational': [
        'A traveler', 'A character', 'A local', 'A visitor', 'A person'
    ]
}

events = {
    'positive': ['discovers something new', 'helps a friend in need', 'finds a lost treasure', 'achieves a lifelong dream','creates something new'],
    'negative': ['loses something valuable', 'encounters a betrayal', 'struggles with a difficult decision', 'abandons their friend', 'is afraid'],
    'neutral': ['attends a meeting', 'observes the surroundings', 'carries a bag', 'speaks to someone', 'is waiting for a call' ]
}   

context = {
    'positive': ['on vacation', 'during christmas time', 'on payday', 'at the zoo', 'on a trip to the park', 'in their home', 'during the summer break'],
    'negative': ['in prison', 'in an abandoned building', 'at a funeral', 'in a time of war', 'during a confrontation', 'in a courtroom', 'in a crime scene'],
    'neutral': ['at the school', 'in a cafeteria', 'during winter', 'in the casino', 'in the office', 'in the hallway', 'at a hostel']
}
```

```python
def generate_story_hints(agents, events, context):
    """
    Generate unique story hints of the form:
    agent + event + context

    Returns a list of dictionaries with:
    - story_hint
    - agent_type
    - event_valence
    - context_valence
    """

    story_hints = []

    for agent_type, agent_list in agents.items():
        for event_valence, event_list in events.items():
            for context_valence, context_list in context.items():

                for agent, event, ctx in itertools.product(
                    agent_list, event_list, context_list
                ):
                    hint = f"{agent} {event} {ctx}"

                    story_hints.append({
                        "story_hint": hint,
                        "agent": agent, 
                        "agent_type": agent_type,
                        "event": event,
                        "event_valence": event_valence,
                        "context": ctx,
                        "context_valence": context_valence
                    })

    return story_hints
```

```python
story_hints = generate_story_hints(agents, events, context)

print(len(story_hints))
random.shuffle(story_hints)
print(story_hints[5])
```

```python
# Save as JSON 
with open('story_hints.json', 'w', encoding='utf-8') as f:
    json.dump(story_hints, f, ensure_ascii=False, indent=4)
```
